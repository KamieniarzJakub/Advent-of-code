```python
data = []

with open("inputs/day1.txt") as file:
    data = file.readlines()

list1 = []
list2 = []

for x in data:
    list1.append(int(x.split()[0]))
    list2.append(int(x.split()[1]))
list1.sort()
list2.sort()
distance = 0
for i in range(len(list1)):
    distance += abs(list1[i]-list2[i])
print("First star:", distance)

dict1 = {}
for x in list1:
    dict1[x] = 0
for y in list2:
    if y in dict1.keys():
        dict1[y]+=1
score = 0
for key, v in dict1.items():
    score+=key*v
print("Second star:", score)
```

```python
data = []

with open("inputs/day2.txt") as file:
    data = file.readlines()
    
def check_levels(values):
    level = 0
    for i in range(len(values)-1):
        diffrence = int(values[i])-int(values[i+1])
        if abs(diffrence)>3:
            return 0
        if diffrence>0:
            level-=1
        elif diffrence<0:
            level+=1
    return abs(level)

very_safe_reports = 0
safe_reports = 0
for line in data:
    values = line.split()
    if check_levels(values)==len(values)-1:
        very_safe_reports+=1
    else:
        n = len(values)
        for i in range(0, n):
            if check_levels(values[0:i]+values[i+1:n])==len(values)-2:
                safe_reports+=1
                break
print("First star:", very_safe_reports)
print("Second star:", very_safe_reports+safe_reports)

example = ['65', '66', '68', '69', '71', '72', '75', '82']
```

# DAY 6

```python
import time
from IPython.display import clear_output
class Direction():
    UP = 0
    RIGHT = 1
    DOWN = 2
    LEFT = 3

def read_data(file_name="day6.txt"):
    mapa = []
    length = 0
    rows = 0
    with open(f"inputs/{file_name}") as file:
        line_length=0
        while True:
            letter = file.read(1)
            if not letter:
                return mapa, length, rows
            if letter!='\n':
                mapa.append(letter)
                line_length+=1
            else:
                length=line_length
                line_length=0
                rows+=1
                
def print_move(mapa, length, rows):
    clear_output(wait=True)
    for i in range(rows+1):
        print(mapa[i*length:(i+1)*length])
    print("\n")
    time.sleep(0.1)

def print_alternative_scenarios(value, total):
    print(f"{value}/{total}")
    clear_output(wait=True)
    return value+1

def calculate_next_move(player, direction, length, rows):
    next_field = player
    if direction==Direction.UP:
        if player//length==0:
            return -1
        next_field = player-length
    elif direction==Direction.RIGHT:
        if player%length==length-1:
            return -1
        next_field = player+1
    elif direction==Direction.DOWN:
        if player//length==rows:
            return -1
        next_field = player+length
    elif direction==Direction.LEFT:
        if player%length==0:
            return -1
        next_field = player-1
    return next_field
    

def move(mapa, length, rows):
    player = mapa.index('^')
    direction = 0
    visited = set()
    while True:
        visited.add(player)
        #print_move(mapa, length, rows)
        next_field = calculate_next_move(player, direction, length, rows)
        if next_field==-1:
            return visited
        if mapa[next_field]=='#':
            direction=(direction+1)%4
            continue
        if mapa[next_field]=='.':
            mapa[next_field]='X'
            
        player=next_field

def alternative_move(mapa, length, rows):
    player = mapa.index('^')
    direction = 0
    visited = set()
    while True:
        visited.add((player, direction))
        next_field = calculate_next_move(player, direction, length, rows)
        if next_field==-1:
            return 0
        if (next_field, direction) in visited:
            return 1
        if mapa[next_field]=='#':
            direction=(direction+1)%4
            continue       
        player=next_field

mapa, length, rows = read_data()
visited = move(mapa, length, rows)
loops=0
i=1
for x in visited:
    i = print_alternative_scenarios(i, len(visited))
    alternative_map = mapa.copy()
    if x!= mapa.index('^'):
        alternative_map[x]='#'
    loops+=alternative_move(alternative_map, length, rows)
print("First star:", len(visited))
print("Second star:", loops)
```

# DAY 7

```python
def read_data(file_name="day7.txt"):
    with open(f"inputs/{file_name}") as file:
        return file.readlines()

def test_2_operators(data):
    total = 0
    possible = []
    for line in data:
        x, *values = line.split()
        x = int(x[:-1])
        possible=[int(values[0])]
        for i in range(1, len(values)):
            temp = []
            for n in possible:
                temp.append(n*int(values[i]))
                temp.append(n+int(values[i]))
            possible = temp.copy()
        if x in possible:
            total+=x
    return total

def test_3_operators(data):
    total = 0
    possible = []
    for line in data:
        x, *values = line.split()
        x = int(x[:-1])
        possible=[int(values[0])]
        for i in range(1, len(values)):
            temp = []
            for n in possible:
                temp.append(n*int(values[i]))
                temp.append(n+int(values[i]))
                temp.append(int(str(n)+values[i]))
            possible = temp.copy()
        if x in possible:
            total+=x
    return total

data = read_data()
print("First star:", test_2_operators(data))
print("Second star:", test_3_operators(data))
```

# DAY 8

```python
def read_data(file_name="day8.txt"):
    with open(f"inputs/{file_name}") as file:
        return file.readlines()
        
def convert_data(data):
    size = [len(data),len(data[0])-1]
    antennas = {}
    for row in range(len(data)):
        for col in range(len(data[row])):
            if data[row][col]!='.' and data[row][col]!='\n':
                if data[row][col] in antennas.keys():
                    antennas[data[row][col]].append((row, col))
                else:
                    antennas[data[row][col]] = [(row, col)]
    return antennas, size

def calculate_antinodes(data, map_size):
    antinodes = [['.']*map_size[1] for _ in range(map_size[0])]
    for antenna in data.keys():
        for i in range(len(data[antenna])-1):
            for j in range(i+1, len(data[antenna])):
                first = data[antenna][i]
                second = data[antenna][j]
                x = first[0] - second[0] 
                y = first[1] - second[1]
                #print(first, x, y)
                if 0<=first[0]+x<map_size[1] and 0<=first[1]+y<map_size[0]:
                    antinodes[first[0]+x][first[1]+y]='#'
                if 0<=second[0]-x<map_size[1] and 0<=second[1]-y<map_size[0]:
                    antinodes[second[0]-x][second[1]-y]='#'
    #print_antinode(antinodes)
    c = 0
    for x in antinodes:
        c+= x.count('#')
    return c

def calculate_T_antinodes(data, map_size):
    antinodes = [['.']*map_size[1] for _ in range(map_size[0])]
    for antenna in data.keys():
        for i in range(len(data[antenna])-1):
            for j in range(i+1, len(data[antenna])):
                first = data[antenna][i]
                second = data[antenna][j]
                x = first[0] - second[0] 
                y = first[1] - second[1]
                #print(first, x, y)
                for k in range(0, map_size[0]):
                    if 0<=first[0]+k*x<map_size[1] and 0<=first[1]+y*k<map_size[0]:
                        antinodes[first[0]+k*x][first[1]+k*y]='#'
                    if 0<=second[0]-x*k<map_size[1] and 0<=second[1]-y*k<map_size[0]:
                        antinodes[second[0]-k*x][second[1]-k*y]='#'
    #print_antinodes(antinodes)
    c = 0
    for x in antinodes:
        c+= map_size[1] - x.count('.')
    return c

def print_antinodes(antinodes):
    for x in antinodes:
        print(*x)
        
                    
data, map_size = convert_data(read_data())
print("First star:", calculate_antinodes(data, map_size))
print("Second star:", calculate_T_antinodes(data, map_size))
```

# DAY 9

```python
from IPython.display import clear_output

def read_data(file_name="day9.txt"):
    with open(f"inputs/{file_name}") as file:
        return file.read()

def convert_data(data):
    disk = []
    id = -1
    for i in range(len(data)):
        if i%2:
            symbol = '.'
        else:
            id+=1
            symbol = id
        for _ in range(int(data[i])):
            disk.append(symbol)
    return disk

def move_files_fragmentation(disk):
    checksum = 0
    j=len(disk)-1
    disk=disk.copy()
    for i in range(len(disk)):
        if disk[i]=='.':
            while j>i and disk[j]=='.':
                j-=1
            if i>=j:
                break
            disk[i], disk[j] = disk[j], disk[i]
        checksum+=i*disk[i]
    return checksum 

def move_files_without_fragmentation(disk):
    i=len(disk)-1
    block_size = 0
    for block_id in range(int(disk[-1]), 0, -1):
        clear_output(wait=True)
        print(block_id)
        while i>=0 and disk[i]!=block_id:
            i-=1
        block_size=0
        while i>=0 and disk[i]==block_id: 
            i-=1
            block_size+=1
        free_space = 0
        for j in range(i+1):
            if disk[j]=='.':
                free_space+=1
                if free_space>=block_size:
                    for k in range(block_size):
                        disk[j-k], disk[i+1+k] = disk[i+1+k], disk[j-k]
                    break
            else:
                free_space=0
    checksum = 0
    for k in range(len(disk)):
        if disk[k]!='.':
            checksum+=k*disk[k]
    return checksum
                    
data = convert_data(read_data())
first = move_files_fragmentation(data)
clear_output(wait=True)
second = move_files_without_fragmentation(data)
print("First star:", first)
print("Second star:", second)
```
